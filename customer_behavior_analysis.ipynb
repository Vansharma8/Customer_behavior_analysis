```python
# Impute missing 'Review Rating' values with the median rating of the respective 'Category'

# groupby + transform = fix values without losing rows

# transform() applies a group-wise operation but keeps the SAME number of rows.

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median())) 
df.isnull().sum()  #verify missing values are handled
```

```python
#rename columns for easier access

df.columns = df.columns.str.replace(' ', '_')
df.columns = df.columns.str.lower()
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

```python
# create a new column for age groups

# qcut divides the data into equal-sized quartiles based on the 'age' column

labels = ['Young Adult', 'Adult', 'Middle Aged', 'Senior']

df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age', 'age_group']].head(10)  
```

```python
# create a new column for purchase frequency in days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
